## Test Stationarity 
A common assumption in many time series techniques is that the data are stationary.

Stationarity	A common assumption in many time series techniques is that the data are stationary.
A stationary process has the property that the mean, variance and autocorrelation structure do not change over time. Stationarity can be defined in precise mathematical terms, but for our purpose we mean a flat looking series, without trend, constant variance over time, a constant autocorrelation structure over time and no periodic fluctuations

## What is a stationary time series 
A stationary time series is one whose properties do not depend on the time at which the series is observed.Time series are stationary if they do not have trend or seasonal effects.
For stationary series statistics  calculated on the time series are consistent over time, like the mean or the variance of the observations.

```python
#check if the mean and the moving variance varies with time 
import warnings 
warnings.filterwarnings('ignore')
from statsmodels.tsa.stattools import adfuller

#yearly moving averages (12 months window)
#window is Size of the moving window. This is the number of observations
#used for calculating the statistic.
def stationary_check(dataset, lags=None): 
    rol_mean =dataset.rolling(12).mean()  #pd.rolling_mean(y, window=12)
    rol_std = dataset.rolling(12).std() #pd.rolling_std(y, window=12)

    #Plot 
    Original_dataset = plt.plot(dataset, color='blue',label='Original')
    
    mean = plt.plot(rol_mean, color='red', label='12-Months Rolling Mean')
    std = plt.plot(rol_std, color='black', label = '12-Months Rolling Std')
    plt.legend(loc='best')
    plt.title('Time series dataset')
    plt.show(block=False)

#lags to specify just the number of values to display
stationary_check(AirPassengers_series, lags=30)
```

Non-stationary time series show seasonal effects, trends, and other structures that depend on the time index. The mean increases with time and thus the series is not stationary. ### Using Summary Statistics to test for stationary 
we can also test stationary time series by reviewing the summary statistics for a dataset for seasons or random partitions and check for obvious or significant differences.
We splited our time series dataset into two (or more) partitions and compare the mean and variance of each group. If they differ and the difference is statistically significant, the time series is likely non-stationary.

Since we are looking at the mean and variance, we can assume that the data conforms to a Gaussian (also called the bell curve or normal) distribution.

```python
AirPassengers_series.hist()
pyplot.show()
```

### Comment on Air Passengers Dataset
The distribution of values does not look like a Gaussian, therefore the mean and variance values are less meaningful.
This squashed distribution of the observations may be another indicator of a non-stationary time series.
I can clearly see the there is no bell curve-like shape of the Gaussian distribution, the seasonality component is growing.
This may suggest an exponential growth from season to season

```python
#split the dataset into two and check for difference in mean and variance 
X = AirPassengers_series.values
split = int(len(X) / 2)
X1, X2 = X[0:split], X[split:]
mean1, mean2 = X1.mean(), X2.mean()
var1, var2 = X1.var(), X2.var()
print('mean1=%f, mean2=%f' % (mean1, mean2))
print('variance1=%f, variance2=%f' % (var1, var2))
```

### Comment on Air Passengers Dataset
I can see the mean and variance look very different, so it is non stationary

## Making Air Passengers Dataset stationary

```python
from numpy import log
#log transform can be used to flatten out exponential change back to a linear relationship.
AirPassengers_series_log = log(AirPassengers_series)
pyplot.hist(AirPassengers_series_log)
pyplot.show()
pyplot.plot(AirPassengers_series_log)
pyplot.show()
```

```python
#halflife is amt of exp decay
AirPassengers_series_expwei_avg = pd.ewma(AirPassengers_series_log, halflife=12)
plt.plot(AirPassengers_series_log)
plt.plot(AirPassengers_series_expwei_avg)
```

### Using EWMA to make the time series stationary

```python
#remove the trend from the time series that is smoothing the time series
log_Ewma = AirPassengers_series_log - AirPassengers_series_expwei_avg
stationary_check(log_Ewma)

pyplot.hist(log_Ewma)
pyplot.show()
```

### Using First-Order Difference to make the time series stationary 

```python
# difference time series
AirPassengers_series_diff = np.diff(AirPassengers_series_log)
plt.plot(AirPassengers_series_diff)
pyplot.show()
pyplot.hist(AirPassengers_series_diff)
pyplot.show()
```
